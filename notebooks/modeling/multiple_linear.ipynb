```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Load data
train_data = pd.read_csv('cleaned_train.csv')
test_data = pd.read_csv('cleaned_test.csv')

# Pilih fitur yang memiliki korelasi dengan target
features = ['Engine Size(L)', 'Cylinders', 'Fuel Consumption City (L/100km)',
            'Fuel Consumption Hwy (L/100km)', 'Fuel Consumption Comb (L/100km)']

X_train = train_data[features]
y_train = train_data['CO2 Emissions(g/km)']

# Inisialisasi model
model = LinearRegression()

# Training model
model.fit(X_train, y_train)

# Lakukan prediksi pada data test
X_test = test_data[features]
predictions = model.predict(X_test)

# Simpan hasil prediksi ke dalam dataframe
results = pd.DataFrame({'Id': range(137142, 195918), 'CO2 Emissions(g/km)': predictions})

# Simpan dataframe ke dalam file CSV
results.to_csv('submission2.csv', index=False)
```

```python
results = pd.read_csv('predictions.csv')

# Bulatkan nilai pada kolom 'CO2 Emissions(g/km)' menjadi satu angka di belakang koma
results['CO2 Emissions(g/km)'] = results['CO2 Emissions(g/km)'].round(1)

# Simpan dataframe ke dalam file CSV
results.to_csv('submission2.csv', index=False)
```
